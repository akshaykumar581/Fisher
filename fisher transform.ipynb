```python
close=df["Close"].tolist()#Created a list with values from close column
```

```python
shi={}#Defining a dictionary
def kebo(a,s):#Defining a function which inserts X value for every close value to insert values to dictionary
  ma=max(a)
  mi=min(a)
  avg=(ma+mi)*0.5
  dif=(ma-avg)
  x=[]
  for i in a:
    x.append((i-avg)/dif)
  shi[s+9]=x[-1]
```

```python
for i in range(244):
  x=[]
  for i in range(i,i+9):
    x.append(close[i])
  kebo(x,i)# Sample random try
```
